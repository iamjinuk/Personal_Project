3. [설치년월] 칼럼을 연도별로 칼럼으로 재정의하기 - (설치년월에서 년도만 따로 추출하기)

```python
df['설치년월'].value_counts()
```

```python
df['설치년도'] = df['설치년월'].map(lambda x : int('20'+(x.split('-')[1])))
df.head()
```
